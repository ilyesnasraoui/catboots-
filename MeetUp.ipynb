```python
data = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
data.isnull().sum(axis = 0)*100/len(data)
```

```python
data= data.replace(np.nan, '', regex=True)
test= test.replace(np.nan, '', regex=True)
```

```python
data.drop(["id"], axis=1, inplace=True)
test.drop(["index","id","price"], axis=1, inplace=True)
```

```python
X = data.drop("price", axis=1)
y = data["price"]
```

```python
cat=data.select_dtypes(include='object').columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
from catboost import Pool, CatBoostRegressor
cb_model = CatBoostRegressor(iterations=25000,
                             learning_rate=0.08,
                             eval_metric='RMSE',
                             random_seed = 123
                             )
cb_model.fit(X_train,y_train,cat,
            eval_set=(X_test,y_test),
             early_stopping_rounds=1000,
             use_best_model=True)
```

```python
y_test_pred = cb_model.predict(test)
predictions = y_test_pred
```

```python
# Set up folds
MAX_ROUNDS = 650
OPTIMIZE_ROUNDS = False
LEARNING_RATE = 0.05
K = 5
kf = KFold(n_splits = K, random_state = 1, shuffle = True)
```

```python
model = CatBoostRegressor(iterations=25000,
                              learning_rate=0.01,
                             eval_metric='RMSE',
                             random_seed = 123
                             )
```

```python
y_valid_pred = 0*y
y_test_pred = 0
X_train, y_train= data.iloc[:,:-1], data.iloc[:,-1]

for i, (train_index, test_index) in enumerate(kf.split(data)):
    
    # Create data for this fold
    y_train, y_valid = y.iloc[train_index], y.iloc[test_index]
    X_train, X_valid = X.iloc[train_index,:], X.iloc[test_index,:]
    print( "\nFold ", i+1)
    
    fit_model = model.fit(X_train, y_train,cat,
                               use_best_model=True,eval_set=(X_valid, y_valid),early_stopping_rounds=600)
        
    # Generate validation predictions for this fold
    pred = fit_model.predict(X_valid,)
    print( "\n==RMSE==:" )
    print(math.sqrt(mean_squared_error(y_valid, pred)))

    y_valid_pred.iloc[test_index] = pred
    y_test_pred += fit_model.predict(test)
    
y_test_pred /= K  # Average test set predictions

print( "\n Final ==RMSE==:" )
math.sqrt(mean_squared_error(y, y_valid_pred))
```
